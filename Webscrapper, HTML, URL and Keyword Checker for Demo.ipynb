```python
def getNaturalText(soup):
    # kill all script and style elements
    for script in soup(["script", "style"]):
        script.extract()    # rip it out

    # get text
    text = soup.get_text()

    # break into lines and remove leading and trailing space on each
    lines = (line.strip() for line in text.splitlines())
    # break multi-headlines into a line each
    chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
    # drop blank lines
    text = '\n'.join(chunk for chunk in chunks if chunk)

    return text
```

```python
def cleanText(text):
    
    from string import punctuation
    text = text.lower().replace(".", ". ")
    text = ''.join(char for char in text if char not in punctuation)
    ##update to include the punctuation particular to online websites e.g TM

    
    return text
```

```python
def getKeywordScores(url="http://www.30agear.com/"):
    
    black_list_score = 0
    white_list_score = 0
    
    stop_words = ['shopify','e-commerce','ecommerce','ekm','woocommerce','bigcommerce','evil','monstrosity']
    good_words = ['recycle','organic','love']
    social_media = ["FB","Facebook","Twitter","Instagram"]
    found_words = []
    
    sample_text = cleanText(getNaturalText(webscrapefromURL(url)['soup']))
                            
    for word in stop_words:
        if word in sample_text and word not in found_words:
            black_list_score += 1
            found_words += word
          
    for word in good_words:
        if word in sample_text and word not in found_words:
            white_list_score += 1 
            found_words += word
    ##could restructure this entire thing to be one Loop
    return (black_list_score,white_list_score)
    
```

```python
def checkURLforEcommerce(url="http://www.30agear.com/"):
    # more keywords to add "Store Shop Outlet Mart Boutique Emporium Bazaar Market Showroom Co-Op Mall Galleria Exchange"
    
    has_ecommerce = False
    store_words= ["store","stores","shop","shopping","shops"]
    
    for word in url.split("/"):
        #the website/store bit
        if word in store_words:
            has_ecommerce= True
            
                                 
    for word in store_words:
        if "." + word in url:
            has_ecommerce= True
            ## the .com bit
    return has_ecommerce
```

```python
def checkHTMLforEcommerce(url="http://www.30agear.com/"):
    has_ecommerce = False
    response = get(url)
    if 'shopify' in response.text:
        has_ecommerce= True
      
    return has_ecommerce
```

```python
def testonURL(url="http://www.30agear.com/"):
   
    return {'HTML_Ecommerce_Check': checkHTMLforEcommerce(url),'URL_Ecommerce_Check':checkURLforEcommerce(url),'Keyword_Ecommerce_Check': getKeywordScores(url)[0]}
```

```python
test_dict = testonURL()
test_dict.values()

#Here, I tested on a shop I know that has Shopify called 30agear.com.
```
